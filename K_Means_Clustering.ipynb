```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv('/content/drive/MyDrive/My Data Science Stuff/Imarticus/for Akmal.zip (Unzipped Files)/Akmal/Marketing.csv')
df.head()
```

```python
df.isna().sum()
```

```python
X=df.iloc[:, [3, 4]]
```

```python
plt.scatter(df.iloc[:, 3], df.iloc[:, 4]);
```

```python
X.head()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans= kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
plt.scatter(df.iloc[:,3], df.iloc[:, 4], c=y_kmeans, cmap='rainbow')
plt.show();
```

```python
plt.scatter(df.iloc[:,3], df.iloc[:, 4], c=y_kmeans, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0],kmeans.cluster_centers_[:, 1], s=150, c='black', label='Centroids')
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1, 10):
  kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

# Another name of wcss is inertia (used it in sklearn)
plt.plot(range(1,10),wcss, color='red')
```
