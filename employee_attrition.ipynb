```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
y=emp_data['Attrition']
```

```python
y= y.map(dict(No=0,Yes=1))
```

```python
data=emp_data
```

```python
data.drop(['EmployeeCount','Over18','StandardHours','EmployeeNumber','Attrition'],axis=1,inplace=True)
```

```python
data.head()
```

```python
X=data
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
```

```python
from sklearn.pipeline import make_pipeline
```

```python
#creating numeric values using onhotencoder via scikitlearn
ohe=OneHotEncoder()
ct= make_column_transformer((ohe,make_column_selector(dtype_include='object')),remainder='passthrough')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
mod1=LogisticRegression()
```

```python
myownpipe= make_pipeline (ct,mod1)
```

```python
myownpipe.fit(X_train,y_train)
```

```python
y_pred=pd.DataFrame(myownpipe.predict(X_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
y_pred_train=pd.DataFrame(myownpipe.predict(X_train))
```

```python
print(classification_report(y_train, y_pred_train))
```
