```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("bank-additional-full.csv",delimiter=';')
data.head()
type(data)
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
c_y = {"y": {"yes": 1, "no": 0}}
data.replace(c_y, inplace=True)
data.head()
```

```python
X= data.iloc[:, data.columns != 'y']
y= data.iloc[:, data.columns == 'y']
```

```python
X.head()
```

```python
y.head()
```

```python
X.describe(include=['object'])
```

```python
X.describe(include=['object']).T
```

```python
X.describe(include=['object']).T.index
```

```python
X.describe(include=['object']).T.index.tolist()
```

```python
categorical=X.describe(include=['object']).T.index.tolist()
```

```python
X=pd.get_dummies(X,columns=categorical,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33, random_state=42)

################################################################################

#Scale the inputs
#from sklearn.preprocessing import StandardScaler
#sc=StandardScaler()

#X_train=sc.fit_transform(X_train)
#X_train=pd.DataFrame(X_train,columns=X_test.columns)

#X_test=sc.transform(X_test)
#X_test=pd.DataFrame(X_test,columns=X_train.columns)
```

```python
from xgboost import XGBClassifier
```

```python
model= XGBClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
